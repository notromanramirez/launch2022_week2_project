```python
#what are draco malfoy's values?
malfoy = """
select concat(student_fname," ", student_lname) as "student", concat(value_1,", ", value_2,", ", value_3,", ", value_4) as "values" from students
inner join house on students.house_id=house.house_id
where student_fname = "draco";

"""

result_malfoy = pd.read_sql(malfoy, con=connection)
result_malfoy

#his values, based on house id, are ambition, cunning, leadership, and resourcefulness 
```

```python
#who is victor krum's favorite class teacher?
viktor = """
select concat(student_fname," ", student_lname) as "student", class_name as 'favorite class', concat(faculty_fname," ", faculty_lname) as 'taught by' from students
join subjects on students.fav_class_id=subjects.class_id
join faculty on subjects.class_name=faculty.subject
where student_fname = "viktor";

"""
result_viktor = pd.read_sql(viktor, con=connection)
result_viktor
```

```python
#top 10 students with the most house points?
house_points = """
SELECT concat(student_fname," ", student_lname) as "student", house_points as 'house points' FROM students
INNER JOIN house on students.house_id = house.house_id
ORDER BY house_points DESC
LIMIT 10;

"""

result_points = pd.read_sql(house_points, con=connection)
result_points

#slytherin is obviously in the lead and all the students with that house are leading in house points. this is followed by soscrofa and ravenclaw 
```

```python
#who's favorite class is located on the field?

field = """
SELECT concat(student_fname," ", student_lname) as "student", class_name as 'favorite class', location FROM students
INNER JOIN subjects ON students.fav_class_id=subjects.class_id
where location="field"
ORDER BY student_fname ASC;
"""

result_field = pd.read_sql(field, con=connection)
result_field
```
