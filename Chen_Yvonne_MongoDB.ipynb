```python
# Import students
students = pd.read_csv('/Users/Michael/W2P/Students')

# Import subjects
subjects = pd.read_csv('/Users/Michael/W2P/Subjects')

# Import house
house = pd.read_csv('/Users/Michael/W2P/House', na_values='NaN')

# Import faculty
faculty = pd.read_csv('/Users/Michael/W2P/Faculty')
```

```python
student_records = students.to_dict('records')
subject_records = subjects.to_dict('records')

#Nest House in Student
for i in range (0, len(student_records)):
    house_id = student_records[i]['house_id']
    house_info = house[house['house_id'] == house_id][['house_name','color_1','color_2','value_1','value_2','value_3','value_4','house_points']].iloc[0].to_dict()
    house_info['house_points'] = int(house_info['house_points'])
    student_records[i]['house_id'] = house_info

# Nest faculty in Subjects
for i in range (0, len(subject_records)):
    faculty_id = subject_records[i]['faculty_id']
    faculty_info = faculty[faculty['faculty_id'] == faculty_id][['faculty_fname','faculty_lname','faculty_school']].iloc[0].to_dict()
    subject_records[i]['faculty_id'] = faculty_info

#Nest Subject in Student
for i in range (0, len(student_records)):
    class_id = student_records[i]['fav_class_id']
    class_info = subject_records[class_id]
    student_records[i]['fav_class_id'] = class_info
```
